## SITUATION PROBLEME : total_ventes()

```python
def total_ventes():
    total = 0

    for heure in range(1,5):
        valeur = int(input(f"Combien d'orange venduesà l'heure {heure} ?"))

        if valeur < 0:
            print("Erreur :une valeur négative n'est pas autorisée. Je mets 0. ")
            valeur = 0
        total += valeur
    return total

resultat = total_ventes()
print("Total d'orange vendues;",resultat)
```

# EXERCICE 1 - Comptage d'erreurs

```python
def compter_erreurs():
    erreur = 0

    for minute in range(1,6):
        nb = int(input(f"Erreurs détectées à la minute {minute} :"))
        if nb < 0:
            nb = 0
        erreur += nb
    return erreur

nb_erreur = compter_erreurs()
print("Total d'erreurs détevtées :",nb_erreur)
```

# EXERCICE 2 - Suivi de la temperature

```python
haut = None
bas = None 

for i in range(1,5):
     t = float(input(f"Temperature {i} ;"))

     if haut is None or t > haut:
         haut = t
     if bas is None or t < bas:
        bas = t

print("Temparature la plus haute :",haut)
print("Temparature la plus basse :",bas)
```

# EXERCICE 3 - Ventes de mobile money

```python
def ventes_mobile_money():
    tot = 0
    for h in range(1,5):
        v = int(input(f"ventes à l'heure {h}:"))

        if v < 0:
            v=0
        tot += v
        return tot

v_mm = ventes_mobile_money()
print("Total vendu ;",v_mm)
```

## EXERCICE 4 - Heures de connexion

```python
def connexion():
    c = 0
    for jour in range(1,4):
        hr = float(input(f"Heures du jour {jour}:"))
        c += hr
    return c

rslt = connexion()

moyenne = rslt / 3
print("Total d'heures :",rslt)
print("Moyenne ;",moyenne)
```

## EXERCICE 6 - Comptage de mots supérieurs à 5 lettres

```python
def compter_mots_longs():
    compteur = 0
    for m in range(1,6):
        mot = input(f"Mot {m} :")

        if len(mot) > 5:
            compteur+=1
    print("Nombre de mots de plus de 5 lettres :",compteur)

compter_mots_longs()
```

## EXERCICE 8 - Simulateur de recharge de téléphone

```python
nombre_de_charge = int(input("Combien de fois avez-vous branché le téléphone ?"))

batterie = 0
for _ in range(nombre_de_charge):
    batterie += 20
    if batterie > 100:
        batterie = 100
print("Niveau final ;",batterie,"%")
```
